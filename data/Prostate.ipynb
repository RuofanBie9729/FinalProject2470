```python
import random
import numpy as np

ind = np.array(['00', '01', '02', '04', '06', '07', '10', '13', '14', 
       '16', '17', '18', '20', '21', '24', '25', '28', '29',
       '31', '32', '34', '35', '37', '38', '39', '40', '41', 
       '42', '43', '44', '46', '47'])
rand_id = random.sample(range(len(ind)), 11)
mask = np.full(len(ind), True, dtype=bool)
mask[rand_id] = False

train_ind = ind[mask]
test_ind = ind[~mask]

print(train_ind)
print(test_ind)
```

```python
import cv2

train_img = np.zeros([256, 256, 1])

for i in train_ind:
    
    print(i)
    
    nii_img  = nib.load('Task05_Prostate/imagesTr/prostate_'+i+'.nii.gz')
    nii_data = nii_img.get_fdata()
    nii_data = nii_data[:, :, :, 0]
    for j in range(nii_data.shape[2]):
        train_img = np.dstack([train_img, cv2.resize(nii_data[:, :, j], (256, 256))])

print(train_img.shape)
```

```python
test_img = np.zeros([256, 256, 1])

for i in test_ind:
    
    print(i)
    
    nii_img  = nib.load('Task05_Prostate/imagesTr/prostate_'+i+'.nii.gz')
    nii_data = nii_img.get_fdata()
    nii_data = nii_data[:, :, :, 0]
    for j in range(nii_data.shape[2]):
        test_img = np.dstack([test_img, cv2.resize(nii_data[:, :, j], (256, 256))])

print(test_img.shape)
```

```python
train_lab = np.zeros([256, 256, 1])

for i in train_ind:
    
    print(i)
    
    nii_lab  = nib.load('Task05_Prostate/labelsTr/prostate_'+i+'.nii.gz')
    nii_data = nii_lab.get_fdata()
    for j in range(nii_data.shape[2]):
        train_lab = np.dstack([train_lab, cv2.resize(nii_data[:, :, j], (256, 256))])
        
print(train_lab.shape)
```

```python
test_lab = np.zeros([256, 256, 1])

for i in test_ind:
    
    print(i)
    
    nii_lab  = nib.load('Task05_Prostate/labelsTr/prostate_'+i+'.nii.gz')
    nii_data = nii_lab.get_fdata()
    for j in range(nii_data.shape[2]):
        test_lab = np.dstack([test_lab, cv2.resize(nii_data[:, :, j], (256, 256))])

print(test_lab.shape)
```

```python
np.save('train_img.npy', train_img[:, :, 1:])
np.save('train_lab.npy', train_lab[:, :, 1:])
train_img_ld = np.load('train_img.npy')
print(train_img_ld.shape)
train_lab_ld = np.load('train_lab.npy')
print(train_lab_ld.shape)
```

```python
np.save('test_img.npy', test_img[:, :, 1:])
np.save('test_lab.npy', test_lab[:, :, 1:])
test_img_ld = np.load('test_img.npy')
print(test_img_ld.shape)
test_lab_ld = np.load('test_lab.npy')
print(test_lab_ld.shape)
```
